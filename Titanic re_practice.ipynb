```python
# Drop cabin column
titanic_data.drop("cabin", axis=1, inplace=True)
```

```python
# Drop all the null values
titanic_data.dropna(inplace=True)
```

```python
titanic_data.head(2)
```

```python
# Convert all the "strings values" to "categorical values" in order to implement logistic regression. 
# Now, convert to "categorical values" into some "dummy variables" using pandas. Logistic regression only take in two values.
# The survive column is  in this case in categorical form.

sex = pd.get_dummies(titanic_data["sex"], drop_first=True)
sex.head(5)
```

```python
# Same applies to the "Embarked" column
embarked = pd.get_dummies(titanic_data["embarked"], drop_first=True)
embarked.head(5)
```

```python
# Same applies to the "pclass" column
plc = pd.get_dummies(titanic_data["pclass"], drop_first=True)
plc.head(5)
```

```python
# Next, concatinate all the three new roles into the dataset
titanic_data=pd.concat([titanic_data, sex, plc, embarked], axis=1)
```

```python
titanic_data.head(5)
```

```python
titanic_data.drop(['sex', 'name', 'passangerid', 'ticket', 'embarked', 'pclass'], axis=1, inplace=True)
titanic_data.head(5)
```

```python
# Train dataset, defining the dependent and independent variables
X = titanic_data.drop("survived", axis=1)
y = titanic_data["survived"]
```

```python
# After defining dependent and independent variables, now, split dataset into training and testing subset

from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Import the LogisticRegression model

from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
# Fit logistic regression model

logmodel.fit(X_train, y_train)
```

```python
# Make predictions

predictions = logmodel.predict(X_test)
```

```python
# Import calssification report

from sklearn.metrics import classification_report
```

```python
classification_report(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```
